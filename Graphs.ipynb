```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from scipy.interpolate import make_interp_spline

file = 'naCL_graph.xlsx'
# Read the Excel file
data = pd.read_excel(file, sheet_name='tg=60nm d1 0.6 d2 1um 1')

# Extract wavelength values
wavelengths = data['wavelength(um)']

# Extract data for different na ranges
na_131_132 = data['na=1.31-1.32']
na_132_133 = data['na=1.32-1.33']
na_133_134 = data['na=1.33-1.34']
na_134_135 = data['na=1.34-1.35']
na_135_136 = data['na=1.35-1.36']
na_136_137 = data['na=1.36-1.37']
na_137_138 = data['na=1.37-1.38']


# Create a function to smooth the data
def smooth(x, y, num_points=300):
    x_new = np.linspace(x.min(), x.max(), num_points)
    spl = make_interp_spline(x, y, k=3)  # Cubic spline interpolation
    y_smooth = spl(x_new)
    return x_new, y_smooth

# Plot the data with smoother lines
plt.figure(figsize=(10, 6))  # Adjust figure size if needed

x_smooth, y_smooth = smooth(wavelengths, na_131_132)
plt.plot(x_smooth, y_smooth, color='b', label='na=1.31-1.32')

x_smooth, y_smooth = smooth(wavelengths, na_132_133)
plt.plot(x_smooth, y_smooth, color='g', label='na=1.32-1.33')

x_smooth, y_smooth = smooth(wavelengths, na_133_134)
plt.plot(x_smooth, y_smooth, color='r', label='na=1.33-1.34')

x_smooth, y_smooth = smooth(wavelengths, na_134_135)
plt.plot(x_smooth, y_smooth, color='c', label='na=1.34-1.35')

x_smooth, y_smooth = smooth(wavelengths, na_135_136)
plt.plot(x_smooth, y_smooth, color='m', label='na=1.35-1.36')

x_smooth, y_smooth = smooth(wavelengths, na_136_137)
plt.plot(x_smooth, y_smooth, color='y', label='na=1.36-1.37')

x_smooth, y_smooth = smooth(wavelengths, na_137_138)
plt.plot(x_smooth, y_smooth, color='k', label='na=1.37-1.38')


# Add labels and legend
plt.xlabel('Wavelength (µm)')
plt.ylabel('Amplitude Sensitivity (RIU$^{-1}$)')
#plt.title('Amplitude Sensitivity vs Wavelength')
plt.legend()

# Show the plot
plt.tight_layout()
plt.show()
```

```python
#d1 variation////////////////
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from scipy.interpolate import make_interp_spline


path = 'wl_CL_AS_tg_variation_graph.xlsx'
data= pd.read_excel(path, sheet_name='CL_d1=0.6,d2=1,p=2,na=1.37')
# Data
wl = data['wl']
d1_1 = data['tg=30nm']
d1_2 = data['tg=40nm']
d1_3 = data['tg=50nm']
d1_4 = data['tg=60nm']


# Smoothing function
def smooth(x, y):
    x_smooth = np.linspace(min(x), max(x), 300)
    y_smooth = make_interp_spline(x, y)(x_smooth)
    return x_smooth, y_smooth

# Specified colors
colors = ['b', 'g', 'r', 'c']

# Plotting
plt.figure(figsize=(8, 6))

# Smoothed plots
for i, (data, label, color) in enumerate(zip([d1_1,d1_2,d1_3,d1_4],
                                             ['tg=30nm', 
                                              'tg=40nm',
                                              'tg=50nm', 
                                              'tg=60nm'], colors)):
    x_smooth, y_smooth = smooth(wl, data)
    plt.plot(x_smooth, y_smooth, label=label, color=color, linewidth=2)

plt.ylabel("Confinement Loss (dB/cm)", fontsize='15')
plt.xlabel("Wavelength (μm)", fontsize='15')
plt.legend(frameon=False, fontsize='15')  # Transparent legend
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.grid(False)  # Remove grid lines completely
#plt.title("Amplitude Sensitivity vs Wavelength")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from scipy.interpolate import make_interp_spline

df= pd.read_excel('phaseMacthing_graph.xlsx')
```

```python
df.isnull().sum().sum()
```

```python
# Drop rows with NaN or infinite values in any column
df = df.replace([np.inf, -np.inf], np.nan).dropna()

# Create a function to smooth the data
def smooth(x, y, num_points=300):
    x_new = np.linspace(x.min(), x.max(), num_points)
    spl = make_interp_spline(x, y, k=3)  # B-spline of order 3
    y_new = spl(x_new)
    return x_new, y_new

# Create the figure and axis
fig, ax1 = plt.subplots(figsize=(8, 6))  # Increased figure size

# Set labels and colors for the primary y-axis
ax1.set_xlabel('Wavelength (µm)', fontsize=14, color='black')
ax1.set_ylabel('Confinement loss (dB/cm)', fontsize=14, color='black')
x_smooth, y_smooth = smooth(df['wl'], df['confinement loss'])
ax1.plot(x_smooth, y_smooth, color='blue', label='Confinement loss')
ax1.tick_params(axis='y', labelcolor='black', labelsize=12)
ax1.tick_params(axis='x', labelcolor='black', labelsize=12)

# Create a secondary y-axis
ax2 = ax1.twinx()

# Set labels and colors for the secondary y-axis
ax2.set_ylabel('Effective Refractive Index', fontsize=14, color='black')
x_smooth_core, y_smooth_core = smooth(df['wl'], df['neff y core real'])
x_smooth_spp, y_smooth_spp = smooth(df['wl'], df['neff spp real'])
ax2.plot(x_smooth_core, y_smooth_core, color='red', label='neff_y_core_(real)')
ax2.plot(x_smooth_spp, y_smooth_spp, color='green', label='neff_y_spp_(real)')
ax2.tick_params(axis='y', labelcolor='black', labelsize=12)

# Remove grid lines
ax1.grid(False)
ax2.grid(False)

# Combine the legends into one
lines1, labels1 = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
ax1.legend(lines1 + lines2, labels1 + labels2, loc="upper right", fontsize=12, bbox_to_anchor=(0.99, 0.99))

# Show the plot
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import r2_score
import matplotlib.pyplot as plt

file = 'linearCorrelation_graph.xlsx'
data = pd.read_excel(file)
# Assuming the Excel file has columns named 'Analyte' and 'Wavelength'
na_values = data['Analyte'].values.reshape((-1, 1))  # Analyte column
resonance_wavelengths = data['Peak Wavelength'].values  # Resonance Wavelength column

# Polynomial Regression (Degree 2)
poly_model = PolynomialFeatures(degree=2)
x_poly = poly_model.fit_transform(na_values)

poly_reg = LinearRegression()
poly_reg.fit(x_poly, resonance_wavelengths)
y_pred_poly = poly_reg.predict(x_poly)
r2_poly = r2_score(resonance_wavelengths, y_pred_poly)

# Extract polynomial coefficients
coeffs = poly_reg.coef_
intercept = poly_reg.intercept_

# Format the equation for display inside the graph
equation_text = f"$R^2 = {r2_poly:.2f}$"
equation_line = f"$y = {coeffs[2]:.2f}x^2 {coeffs[1]:.2f}x {'+' if intercept > 0 else '-'} {(intercept):.2f}$"

# Plotting Polynomial Regression Results
plt.figure(figsize=(6, 4))
plt.scatter(na_values, resonance_wavelengths, color="blue", label="Data Points")
plt.plot(na_values, y_pred_poly, color="orange", label="Polynomial Regression")


# Adjust the text positions dynamically to keep them **inside** the graph
x_text_pos = min(na_values) + 0.00
y_text_pos = min(resonance_wavelengths) + (max(resonance_wavelengths) - min(resonance_wavelengths)) * 0.2

plt.xticks(fontsize=12)
plt.yticks(fontsize=12)

plt.text(x_text_pos, y_text_pos + 0.08, equation_text, fontsize=13, color='black', ha='left')
plt.text(x_text_pos, y_text_pos + 0.05, equation_line, fontsize=13, color='black', ha='left')

# Labels and Titles
plt.xlabel("Analyte Refractive Index (na)", fontsize=12)
plt.ylabel("Resonance Wavelength (µm)", fontsize=12)
#plt.title("Polynomial Regression for Resonance Wavelength")
plt.legend(fontsize='12')
plt.grid(True, linestyle="--", alpha=0.6)
plt.tight_layout()
plt.show()
```
